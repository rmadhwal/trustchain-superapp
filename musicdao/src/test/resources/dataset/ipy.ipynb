```python
import numpy as np
import os
from pathlib import Path

x_values = []
y_values = []
reputation_gains = []
top_100 = []
top_1000 = []
top_2000 = []
top_5000 = []

with open('sybil_experiment_out_flat_2.txt', "r") as of:
    lines = of.readlines()
    i = 0
for line in lines:
    if i == 0:
        words = line.split()
        beta_decay = float(words[2])
        exploration = float(words[4])
        reputation_gain = float(words[-1])
        x_values.append(beta_decay)
        y_values.append(exploration)
        reputation_gains.append(reputation_gain)
        i = i + 1
    elif i == 1:
        top_100.append(int(line.split()[-1]))
        i = i + 1
    elif i == 2:
        top_1000.append(int(line.split()[-1]))
        i = i + 1
    elif i == 3:
        top_2000.append(int(line.split()[-1]))
        i = i + 1
    elif i == 4:
        top_5000.append(int(line.split()[-1]))
        i = 0
```
