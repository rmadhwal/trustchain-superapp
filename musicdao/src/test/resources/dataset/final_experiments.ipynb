```python
import numpy as np
import os
from pathlib import Path

song_score_og_values = []
top_100_og_values = []
top_1000_og_values = []
top_10000_og_values = []
top_100000_og_values = []
song_score_twitter_values = []
top_100_twitter_values = []
top_1000_twitter_values = []
top_10000_twitter_values = []
top_100000_twitter_values = []
alpha_decay_values = []

with open('recommendations_quality.txt', "r") as of:
    lines = of.readlines()
    i = 0
for line in lines:
    if i == 0:
        words = line.split()
        alpha_decay = float(words[2])
        alpha_decay_values.append(alpha_decay)
        i = i + 1
    elif i == 1:
        words = line.split()
        score = float(words[-1])
        song_score_og_values.append(score)
        i = i + 1
    elif i == 2:
        top_100_og_values.append(float(line.split()[-1]))
        i = i + 1
    elif i == 4:
        top_1000_og_values.append(float(line.split()[-1]))
        i = i + 1
    elif i == 5:
        top_10000_og_values.append(float(line.split()[-1]))
        i = i + 1
    elif i == 6:
        top_100000_og_values.append(float(line.split()[-1]))
        i = 0
    else:
        i = i + 1
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
category = ['100', '1000', '10000', '100000']
percentage = [top_100_og_values[2], top_1000_og_values[2], top_10000_og_values[2], top_100000_og_values[2]]
ax.bar(category,percentage)
ax.set_title('Percentage of sampled node where missing song is recommended')
ax.set_xlabel('Top X recommendations (Alpha Decay: 0.1)')
plt.show()
```

```python
alpha_decays = []
reputation_gains = []
top_100 = []
top_1000 = []
top_2000 = []
top_5000 = []

with open('giga_sybil_experiment_alpha_final.txt', "r") as of:
    lines = of.readlines()
    i = 0
for line in lines:
    if i == 0:
        words = line.split()
        alpha_decay = float(words[2])
        reputation_gain = float(words[-1])
        alpha_decays.append(alpha_decay)
        reputation_gains.append(reputation_gain)
        i = i + 1
    elif i == 1:
        top_100.append(int(line.split()[-1]))
        i = i + 1
    elif i == 2:
        top_1000.append(int(line.split()[-1]))
        i = i + 1
    elif i == 3:
        top_2000.append(int(line.split()[-1]))
        i = i + 1
    elif i == 4:
        top_5000.append(int(line.split()[-1]))
        i = 0
```

```python
alpha_decay_values_str = [str(h) for h in alpha_decays]
plt.figure(figsize=(10, 6))
plt.bar(alpha_decay_values_str, reputation_gains)
plt.xlabel('Alpha Decays')
plt.ylabel('Combined Ranking Score of Sybil Songs')
plt.title('Ranking Score for Sybil songs in Giga Sybil Attack with increasing Alpha Decay')
plt.show()
```

```python
index = np.arange(len(alpha_decays))
bar_width = 0.35
plt.figure(figsize=(10, 6))
plt.plot(index, top_100, color = 'r', label="Top 100 Recommendations")
plt.xlabel('Alpha Decays')
plt.ylabel('Ranking Significance Score')
plt.title('Sybils in Top 100 recommendations with increasing alpha decay in Giga Sybil Attack')
plt.xticks(index + bar_width/2, alpha_decay_values_str)
plt.legend()
plt.show()
```

```python
index = np.arange(len(alpha_decays))
bar_width = 0.35
plt.figure(figsize=(10, 6))
plt.plot(index, top_1000, color = 'g', label="Top 1000 Recommendations")
plt.xlabel('Alpha Decays')
plt.ylabel('Ranking Significance Score')
plt.title('Sybils in Top 1000 recommendations with increasing alpha decay in Giga Sybil Attack')
plt.xticks(index + bar_width/2, alpha_decay_values_str)
plt.legend()
plt.show()
```
